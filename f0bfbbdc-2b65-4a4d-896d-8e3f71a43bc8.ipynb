### Названия, или метки, столбцов датафрейма

### Типы данных

В столбце user_score найдены значения tbd. Фактически это пропущенное значение, которое, скорее всего, означает нехватку голосов для определения рейтинга, поэтому они были переведены в NaN:

```python
df.info()
```

### Наличие пропусков в данных

Посчитаем количество пропусков в каждом столбце в абсолютных и относительных значениях:

```python
absolute = df.isna().sum()
relative = round(df.isna().mean() * 100, 2)

print('Количество пропусков:\n', absolute)
print('Доля пропусков:\n', relative)
```

name и genre - пропущено всего по 2 значения (0.01%). Это случайные единичные пропуски, их можно удалить:

```python
# удаляем строки где пропущены имя игры или жанр
df = df.dropna(subset=['name', 'genre'])
```

year_of_release - пропущено 275 значений (1.62%). У некоторых старых или редких игр год не был внесен в базу. Эти строки также можно удалить, тк наш дальнейший анализ привязан к временному отрезку 2000–2013 гг. Также преобразуем год выхода в целые числа int:

```python
# удаляем строки где пропущен год выхода
df = df.dropna(subset=['year_of_release'])

# переводим года выпуска в обычные целые числа int
df['year_of_release'] = df['year_of_release'].astype(int)
```

eu_sales (6 пропусков) и jp_sales (4 пропуска) - возникли после очистки unknown. Можно заполнить средними значениями в зависимости от платформы и года выпуска:

```python
# ЗАПОЛНЯЕМ ПРОПУСКИ В ЕВРОПЕЙСКИХ ПРОДАЖАХ

# получаем список строк с пропусками в столбце eu_sales
eu_rows = df[df['eu_sales'].isna()].index

for i in eu_rows:
    # получаем платформу и год для строки с пропуском
    platform = df.loc[i, 'platform']
    year = df.loc[i, 'year_of_release']
    
    # выбираем игры с такой же платформой и годом
    games = df[(df['platform'] == platform) & (df['year_of_release'] == year)]
    
    # вычисляем среднее значение продаж
    mean = games['eu_sales'].mean()
    
    # заменяем пропуски на среднее значение
    df.loc[i, 'eu_sales'] = mean
    
# ЗАПОЛНЯЕМ ПРОПУСКИ В ЯПОНСКИХ ПРОДАЖАХ

eu_rows = df[df['jp_sales'].isna()].index

for i in eu_rows:
    platform = df.loc[i, 'platform']
    year = df.loc[i, 'year_of_release']
    
    games = df[(df['platform'] == platform) & (df['year_of_release'] == year)]
    
    mean = games['jp_sales'].mean()
    
    df.loc[i, 'jp_sales'] = mean
```

critic_score (51.39%) и user_score (54.66%) - пропущено более половины данных. Заполнять их искусственно (средним или медианой) нельзя, это исказит распределение оценок. Оставим пропуски как есть (NaN), чтобы не вносить погрешность.

rating - пропущено 40.52% данных. Некоторые игры не получили рейтинг от американской копмании ESRB. Заменим пропуски на строковое значение-индикатор unknown:

```python
# заполняем пропущенный возрастной рейтинг индикатором unknown
df['rating'] = df['rating'].fillna('unknown')
```

**Возможные причины возникновения пропусков**
1. name и genre - MCAR (полностью случайные пропуски). Эти пропуски единичны (всего по 2 значения на всю таблицу). Они возникли изза случайной технической ошибки при копировании данных или человеческого фактора (оператор случайно пропустил поле при ручном вводе).
2. year_of_release - MCAR / MAR (полностью случайные или случайные). Пропуски составляют 1.62%. Часть из них - технические ошибки при переносе или слиянии баз данных разных игровых платформ (MCAR). Другая часть связана с тем, что для старых, редких или малоизвестных игр точный год официального релиза мог быть не зафиксирован на момент составления таблицы (MAR).
3. eu_sales и jp_sales - MCAR (полностью случайные пропуски). Образовались в результате предобработки, когда мы принудительно заменили некорректные текстовые заглушки unknown на числовой формат NaN. Появление unknown в исходном файле вызвано отсутствием информации о продажах на момент выгрузки данных.
4. critic_score и user_score - MAR / MNAR (случайные или неслучайные). Пропущено более половины всех данных (51.39% и 54.66%). У большого количества ретро-игр ранее не существовало оценок в цифровом виде. Некоторые малобюджетные игры могут вообще не получать обзоров от СМИ и критиков. В случае с пользователями часть пропусков - это бывшие текстовые значения tbd (будет определено). Рейтинг у этих игр отсутствует физически, так как пользователи оставили мало отзывов, и система не смогла рассчитать средний балл.
5. rating - MNAR (неслучайные пропуски). Пропущено 40.52% данных. Это связано со спецификой признака. Организация ESRB, присваивающая данный рейтинг, является американской. Игры, выпущенные до её основания (1994 год) или издававшиеся исключительно для внутренних рынков Японии и Европы, этот рейтинг не получали.

## Фильтрация данных

На основе очищенных данных сформируем новый актуальных срез df_actual, в который войдет временной отрезок с 2000 по 2013 год включительно:

```python
# фильтруем данные
df_actual = df[(df['year_of_release'] >= 2000) & (df['year_of_release'] <= 2013)].copy()

# проверка фильтрации
print("Минимальный год:", df_actual['year_of_release'].min())
print("Максимальный год:", df_actual['year_of_release'].max())

# подсчитаем количество строк после фильтрации
print("Количество строк в актуальном срезе:", len(df_actual))
```

## Категоризация данных

Проведем категоризацию по оценкам пользователей (user_score), разобъем все игры на категории:
- высокая оценка - от 8 до 10 включительно;
- средняя оценка - от 3 до 8, не включая правую границу интервала;
- низкая оценка - от 0 до 3, не включая правую границу интервала.

```python
df_actual['user_category'] = pd.cut(
    df_actual['user_score'],
    bins = [0, 3, 8, 10.01],
    labels = ['низкая оценка', 'средняя оценка', 'высокая оценка'],
    right = False
)
```

Теперь проведем категоризацию по оценкам критиков (critic_score), выделим такие категории:
- высокая оценка - от 80 до 100 включительно;
- средняя оценка - от 30 до 80, не включая правую границу интервала;
- низкая оценка - от 0 до 30, не включая правую границу интервала).

```python
df_actual['critic_category'] = pd.cut(
    df_actual['critic_score'],
    bins = [0, 30, 80, 100.01],
    labels = ['низкая оценка', 'средняя оценка', 'высокая оценка'],
    right = False
)
```

Проверим результат - сгруппируем данные по выделенным категориям и посчитаем количество игр в каждой категории:

```python
user = df_actual.groupby('user_category')['name'].count()
print('Распределение категорий по оценкам пользователей:\n', user)

critic = df_actual.groupby('critic_category')['name'].count()
print('Распределение категорий по оценкам критиков:\n', critic)
```

Также определим 7 самых популярных игровых платформ по количеству игр, выпущенных за весь актуальный период:

```python
top7_platforms = df_actual.groupby(['platform'])['name'].count().sort_values(ascending=False).head(7)

print('Топ-7 платформ по количеству игр, выпущенных в 2000-2013 гг.:\n', top7_platforms)
```
